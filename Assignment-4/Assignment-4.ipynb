```python
driver=webdriver.Chrome(r'C:\Users\User\Desktop\veeresh\Internship(test)/chromedriver.exe')
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
video_name=[]
artist=[]
rank=[]
date=[]
views=[]

video=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for i in video:
    video_name.append(i.text.replace("[*]",""))
    
artist_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for i in artist_tag:
    artist.append(i.text)

rank_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in rank_tag:
    rank.append(i.text)

dt_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
for i in dt_tag:
    date.append(i.text)

view=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for i in view:
    views.append(i.text)
```

```python
len(video_name),len(artist),len(rank),len(date),len(views)
```
