```python
det = []
ratings = []
prp = []
fcl10 = []
des10 = []

scrpdet10 = driver.find_elements(By.XPATH,'//div[@class="property-name"]')
for s in scrpdet10:
    det.append(s.text.split('\n'))
scrrtn10 = driver.find_elements(By.XPATH,'//div[@class="score"]')
for r in scrrtn10:
    ratings.append(r.text.split('\n')) 
prp10 = driver.find_elements(By.XPATH,'//strong[@class="current"]')
for p in prp10:
    prp.append(p.text.split('\n')) 
fcl = driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
for f in fcl:
    fcl10.append(f.text.split('\n')) 
ds = driver.find_elements(By.XPATH,'//div[@class="keyword"]')
for d in ds:
    des10.append(d.text.split('\n')) 
    time.sleep(3)
```
